# Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Load Dataset

```python
train_numeric = pd.read_csv("/content/drive/MyDrive/train_numeric.csv")
test_numeric = pd.read_csv("/content/drive/MyDrive/test_numeric.csv")
```

```python
# Function to plot histograms for numeric columns
def plot_histograms(df, title):
    df.hist(bins=20, figsize=(15, 10), layout=(3, 3), grid=False, color='skyblue', edgecolor='black')
    plt.suptitle(title, fontsize=16)
    plt.subplots_adjust(top=0.92, hspace=0.3, wspace=0.3)

    # Add labels to each subplot
    for ax in plt.gcf().axes:
        ax.set_xlabel(ax.get_title(), fontsize=12)
        ax.set_title("")
        ax.set_ylabel("Frequency", fontsize=12)
```

```python
# Plot histograms for numeric columns in the train dataset
plot_histograms(train_numeric, "Histograms for Numeric Columns in Train Dataset")
plt.show()
```

```python
# Plot histograms for numeric columns in the test dataset
plot_histograms(test_numeric, "Histograms for Numeric Columns in Test Dataset")
plt.show()
```
